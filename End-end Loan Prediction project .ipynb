## Building Evaluation Function

```python
# create function
def cal_auc_score(y_test, y_pred_proba):
    """
    calculate the Area Under the ROC curve (AUC) score
    
    """
    # calculate the AUC
    auc_score = roc_auc_score(y_test, y_pred_proba)
    print(f"AUC Scores is: {auc_score}.4f")
    return auc_score



# let me create functions for other evaluation metrix

def show_scores(clf):
    # Make predictions
    train_preds = clf.predict(X_train)
    valid_preds = clf.predict(X_valid)
    
    # Store scores in a dictionary
    scores = {
        "Train Confusion Matrix": confusion_matrix(y_train, train_preds),
        "Valid Confusion Matrix": confusion_matrix(y_valid, valid_preds),
        "Train Classification Report": classification_report(y_train, train_preds),
        "Valid Classification Report": classification_report(y_valid, valid_preds)
    }
    
    # Return the scores dictionary
    return scores


```

```python
clf =RandomForestClassifier(n_jobs=-1,random_state=42, max_samples=10000)
clf.fit(X_train, y_train)
clf.get_params()
```

```python
# Call the function and store the results in clf_scores
clf_scores = show_scores(clf)

# Print each score
for key, value in clf_scores.items():
    print(f"{key}:\n{value}\n")
```

## Hyperparameters tuning with RandomizedSearchCV

```python
from sklearn.model_selection import RandomizedSearchCV, GridSearchCV
# different RandomForestClassifier Hyperparameter 
rfc_grid = {
    "n_estimators": np.arange(20,200,20),
    "max_features":["auto","sqrt","log2"],
    "max_depth": [None, 10,20,30,40,50],
    "min_samples_split":[2,5,10],
    "min_samples_leaf":[1,2,4],
    "bootstrap":[True, False],
    "criterion":["gini","entropy"],
    "max_samples":[10000]
}
# instantiating the randomizedsearchsv model
rf_clf =RandomizedSearchCV(RandomForestClassifier(n_jobs=-1,random_state=42),
                          param_distributions=rfc_grid,
                          n_iter=50,
                          cv=5,
                          verbose=True)
rf_clf.fit(X_train, y_train)
```
